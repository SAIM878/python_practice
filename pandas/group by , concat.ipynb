```python
india_weather = {
    'city':['kolkata','mumbai','dehli'],
    'temperature': [92,96,90],
    'humidity':[80,60,78]
    
}
```

```python
india_weather = pd.DataFrame(india_weather)
india_weather
```

```python
us_weather = {
    'city':['nyc','miami','chicago'],
    'temperature': [21,14,40],
    'humidity':[88,50,70]
}
us_weather =pd.DataFrame(us_weather)
us_weather
```

```python
df = pd.concat([india_weather,us_weather])
df
```

```python
df = pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
df = pd.concat([india_weather,us_weather],keys=["india","us"])
df
```

```python
df.loc["us"]
```

```python
df1 = pd.DataFrame({'city':['kolkata','mumbai','dehli'],
 'temperature': [92,96,90]})
df1
```

```python
df2 = pd.DataFrame({
    'city':['kolkata','mumbai','dehli'],
    'humidity':[88,50,70]
    })
df2
```

```python
df3 = pd.merge(df1,df2,on="city")
df3
```

```python
df1 = pd.DataFrame({'city':['kolkata','mumbai','dehli','islamabad'],
 'temperature': [92,96,90,10]})
df1
```

```python
df2 = pd.DataFrame({
    'city':['kolkata','mumbai','dehli'],
    'humidity':[88,50,70]
    })
df2
```

```python
df3 = pd.merge(df1,df2,on='city',how="outer")
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how="outer",indicator=True)
df3
```
